## Check the presence of outliers

```python
plt.figure(figsize=(40,10))
plt.subplot(2,2,1)
sns.boxplot(x=df1['Sales'])
plt.subplot(2,2,2)
sns.boxplot(x=df1['Quantity'])
plt.subplot(2,2,3)
sns.boxplot(x=df1['Discount'])
plt.subplot(2,2,4)
sns.boxplot(x=df1['Profit'])
```

### Inference: Outliers are present in Sales,Quantity,Discount and Profit.

```python
#Removing outliers
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
low,upp=remove_outlier(df1['Sales'])
df1['Sales']=np.where(df1['Sales']>upp,upp,df1['Sales'])
df1['Sales']=np.where(df1['Sales']<low,low,df1['Sales'])
```

```python
low,upp=remove_outlier(df1['Quantity'])
df1['Quantity']=np.where(df1['Quantity']>upp,upp,df1['Quantity'])
df1['Quantity']=np.where(df1['Quantity']<low,low,df1['Quantity'])
```

```python
low,upp=remove_outlier(df1['Discount'])
df1['Discount']=np.where(df1['Discount']>upp,upp,df1['Discount'])
df1['Discount']=np.where(df1['Discount']<low,low,df1['Discount'])
```

```python
low,upp=remove_outlier(df1['Profit'])
df1['Profit']=np.where(df1['Profit']>upp,upp,df1['Profit'])
df1['Profit']=np.where(df1['Profit']<low,low,df1['Profit'])
```

```python
plt.figure(figsize=(40,10))
plt.subplot(2,2,1)
sns.boxplot(x=df1['Sales'])
plt.subplot(2,2,2)
sns.boxplot(x=df1['Quantity'])
plt.subplot(2,2,3)
sns.boxplot(x=df1['Discount'])
plt.subplot(2,2,4)
sns.boxplot(x=df1['Profit'])
```
